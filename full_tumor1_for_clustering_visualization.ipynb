# Notes for Following the Clustering Tutorial

### Data loading

Import the data I give you like this:

```adata = sc.read_h5ad('full_tumor1.h5ad')```

_Don't use the 10X matrix files in tutorial line # 5_

### Data processing

1. Use Leiden not Louvain 
2. Beware that dataset-specific examples won't work, like:

```sc.pl.umap(adata, color=['CST3', 'NKG7', 'PPBP'])```

### Visualization

Show the cell type (original cluster #) and the leiden cluster in the UMAP plots, not the genes that the tutorial uses

```sc.pl.umap(adata, color=['leiden','cell_type'])```

* Note: cell_types are numbers here, unlike actual names in the Seurat example
