# Notes for Following the PAGA Tutorial

### Data loading

```adata = sc.read_h5ad('tumor1.h5ad')```

_Don't pull the Paul 15 data: adata = sc.datasets.paul15()_

### Data processing

1. Use Leiden instead of Louvain
2. Try 3 resolutions (1.0,  0.5,  0.1) here:

```sc.tl.leiden(adata, resolution=1.0)``` 

3. Move forward with the lowest resolution. 

4. Don't annotate yet

skip this: ```adata.obs['louvain_anno'] = adata.obs['louvain']```

```sc.pl.paga(adata, color=['cell_type','leiden'])```
